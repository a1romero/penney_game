```python
def shuffle_deck():
    rng = np.random.default_rng()
    deck = np.ndarray.flatten((np.stack((np.ones(26), np.zeros(26)), axis= 0).astype(int)))
    rng.shuffle(deck)
    return deck, ''.join(map(str, deck))
```

```python
def sum_games():
    data = 'data/'
    files = [file for file in os.listdir(data)]
    games_total = None
    for file in files:
        file_path = os.path.join(data,file)
        game = np.load(file_path, allow_pickle=True)

        if games_total is None:
            games_total = game
        else:
            games_total += game
    num_games = len(files)
    return np.divide(games_total, num_games)
```
